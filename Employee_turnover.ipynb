```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Sample data for demonstration
data = pd.Series([10, 12, 15, 18, 20, 22, 25, 28, 30, 32, 35, 38, 40, 42, 45, 48, 50, 52, 55, 58, 60, 62, 65, 68])

# Splitting data into train and test sets
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

# Fitting ARIMA model
model = ARIMA(train, order=(1,1,1))  # ARIMA(p,d,q) order
model_fit = model.fit()

# Forecasting
forecast = model_fit.forecast(steps=len(test))

# Calculating RMSE
mse = mean_squared_error(test, forecast)
rmse = np.sqrt(mse)
print('Test RMSE: %.3f' % rmse)

# Plotting results
plt.plot(test.index, test.values, label='Actual')
plt.plot(test.index, forecast, color='red', label='Forecast')
plt.legend()
plt.title('Employee Turnover Forecasting')
plt.xlabel('Time')
plt.ylabel('Turnover Rate')
plt.show()
```
